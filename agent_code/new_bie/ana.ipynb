```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
reward_log_path = 'logs/reward_log.csv'

data = pd.read_csv(reward_log_path)

plt.style.use('ggplot')
plt.figure(figsize=(10, 5))
plt.plot(data['episode'], data['reward'], label='Reward per Round', color='blue')
plt.title('Reward per Round Over Time')
```
